```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import folium
from scipy import stats
from potosnail import Wrappers

wr = Wrappers()

EDA =  pd.read_csv('https://raw.githubusercontent.com/spe301/ML-vs-DL/main/data/Tanzania_Water_Pumps/train_df_after_EDA.csv')
EDA.head()
```

# EDA Questions
1. Which major funder, top 100 most common, has highest percentage of nonfunctional pumps?
2. is the difference in water pump functionality statisticy signifigant between commercially managed and group managed pumps?
3. What is the most reliable water source, this is measured by percentage of pumps that are fully functional and have 'good' water.
4. display the status of the pumps in geographic format.

```python
funders = dict(EDA['funder'].value_counts()) #turning value counts into a dictionary so it's easier to work with
funders_df = pd.DataFrame(list(funders.keys())) #turning this into a pandas dataframe because pandas is faster than raw python
funders_df.columns = ['funder'] 
funders_df['count'] = list(funders.values())
funders_df = funders_df[:100] #taking the top 100
funders_list = list(funders_df['funder'])
```

```python
percentages = {}
for funder in funders_list:
  sample = EDA.loc[EDA['funder']==funder]
  defective = sample.loc[sample['label']==0]
  percentages[round((len(defective)/len(sample)), 2) * 100] = funder

top5 = []
for element in sorted(percentages)[-5:]:
  top5.append(percentages[element])

x = top5
y = sorted(percentages)[-5:]
fig, ax = plt.subplots(figsize=(15, 10))
sns.barplot(x, y, ax=ax)
plt.xlabel('Funder')
plt.ylabel('% Of Pumps Defective')
plt.rcParams['font.family'] = 'MS PMincho'
```

Conclusion: The vast majority of the water pumps established by these funders are defective. As far as context goes in regards to these funders all I could find was the Halamashauri Ya Wilaya Sikonge website http://www.sikongedc.go.tz/.

Reccomendation: We should allocate repair workers towards pumps established by these funders. When a new pump is funded by one of these orginizations we can predict a good chance that the pump may not work properly.

```python
commercial = list(EDA.loc[EDA['management_group']=='commercial']['label'].sample(100))
group = list(EDA.loc[EDA['management_group']=='user-group']['label'].sample(100))
stats.ttest_ind(commercial, group)
```

```python
sources = list(dict(EDA['source'].value_counts()).keys())

percentages = []
for source in sources:
  sample = EDA.loc[EDA['source']==source]
  subsample = sample.loc[sample['quality_group']=='good']
  percentages.append((len(subsample)/len(sample))*100)

fig, ax = plt.subplots(figsize=(15, 10))
sns.barplot(sources, percentages, ax=ax)
plt.xlabel('Source')
plt.ylabel('% Of Pumps with good quality water')
plt.rcParams['font.family'] = 'MS PMincho'
```

Conclusion: Springs, Rivers, and Rain Harvesting are very reliable water sources, while machine dbh along with unknown methods don't seem to do to great. For context, a dbh machine extracts water vapor from the air and turns it into potable water.

Reccomendation: Villages should use proven methods for water collection and avoid using dbh machines unless absolutley nessecary. Villages should only use dbh machines in desert-like areas and without any body of water.
